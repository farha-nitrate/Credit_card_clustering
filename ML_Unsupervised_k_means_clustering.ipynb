```python
from scipy.stats import zscore
```

```python
z_scores = df.apply(zscore)

threshold = 3

# Identify outliers 
outliers = (z_scores.abs() > threshold)

# Remove rows with outliers
df_no_outliers = df[~(outliers.any(axis=1))]

# Replace outliers with NaN
df_replaced_outliers = df.mask(outliers)

# replacing outliers with the median of each column
df_median_replaced_outliers = df.apply(lambda x: np.where(outliers[x.name], x.median(), x))
```

# Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve

```python
from sklearn.cluster import KMeans
```

```python
# Calculate WCSS for different values of k
wcss = []
k_range = range(1,11)  # Test k values from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)  # WCSS for the current k

# Step 3: Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, wcss, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.show()
```

# Q5. Find the optimum value of k for k-means clustering using the silhouette score method. Build a K-means clustering model and specify the number of observations in each cluster using a bar plot

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
sil_scores=[]
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(df2)
    score=silhouette_score(df2, kmeans.labels_)
    sil_scores.append(score)
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), sil_scores, marker='o')
plt.title('Silhouette score method ')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
optimal_k = range(2, 11)[sil_scores.index(max(sil_scores))]
print(f"The optimal value of k is: {optimal_k}")
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(df2)
cluster_counts = pd.Series(kmeans.labels_).value_counts().sort_index()
plt.figure(figsize=(8, 6))
cluster_counts.plot(kind='bar', color='skyblue') 
plt.title(f'Number of Observations in Each Cluster (k={optimal_k})')
plt.xlabel('Cluster')
plt.ylabel('Number of Observations')
plt.xticks(rotation=0)
plt.show()
```
